```python
#Converting the datatype of Review column - String / object. And replacing or removing special \
# characters in Review column.
df['Review'].astype(str)
df['Review'] = df['Review'].str.replace(r'[!,@,#]','',regex=True)
df.head(20)
```

```python
# Converting the review column values to lower-case.
df['Review'] = df['Review'].str.lower()
df['Review']
```

```python
# Checking if there is any null or Nam values present in dataset. if present print count number of rows
missing_val = df.isnull().sum()
missing_val
```

```python
# Display or print whole dataset which contains null or Nan values.
rows_with_null = df[df.isnull().any(axis=1)]
rows_with_null
```

```python
# Print the Review Column values which contains null or Nan.
null_values = df['Review'][df['Review'].isnull()]
null_values
```

```python
# Deleting the whole null / Nan values from dataset. For structured Dataset.
df = df.dropna()
df
```
